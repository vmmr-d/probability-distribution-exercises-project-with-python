# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
# TODO
from scipy.stats import norm, uniform

# Funciones para la Distribución Normal

# Cálculo de la función de Densidad de Probabilidad (PDF) de la Distribución Normal.
def dnormal(x, mean=0, sd=1):
    return norm.pdf(x, loc=mean, scale=sd)

# Cálculo de la función de Distribución Acumulativa (CDF) de la Distribución Normal.
def pnormal(q, mean=0, sd=1):
    return norm.cdf(q, loc=mean, scale=sd)

# Cálculo de la función cuantil (inversa de la CDF) de la Distribución Normal.
def qnormal(p, mean=0, sd=1):
    return norm.ppf(p, loc=mean, scale=sd)

# Cálculo que genera Variables Aleatorias a partir de la Distribución Normal.
def rnormal(n, mean=0, sd=1):
    return norm.rvs(loc=mean, scale=sd, size=n)

# Funciones para la Distribución Uniforme

# Cálculo de la función de Densidad de Probabilidad (PDF) de la Distribución Uniforme.
def duniforme(x, min=0, max=1):
    return uniform.pdf(x, loc=min, scale=max-min)

# Cálculo de la función de Distribución Acumulativa (CDF) de la Distribución Uniforme.
def puniforme(q, min=0, max=1):
    return uniform.cdf(q, loc=min, scale=max-min)

# Cálculo de la función cuantil (inversa de la CDF) de la Distribución Uniforme.
def quniforme(p, min=0, max=1):
    return uniform.ppf(p, loc=min, scale=max-min)

# Cálculo que genera Variables Aleatorias a partir de la Distribución Uniforme.
def runiforme(n, min=0, max=1):
    return uniform.rvs(loc=min, scale=max-min, size=n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Parámetros
size = 1000
mean = 0
sd = 1
min_val = 0
max_val = 1

# Generar datos
data_normal = rnormal(size, mean=mean, sd=sd)
data_uniforme = runiforme(size, min=min_val, max=max_val)

# Calcular funciones de densidad
x_norm = np.linspace(mean - 3*sd, mean + 3*sd, 100)
y_norm = dnormal(x_norm, mean=mean, sd=sd)

x_unif = np.linspace(min_val, max_val, 100)
y_unif = duniforme(x_unif, min=min_val, max=max_val)

# Crear subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Gráfico para la distribución normal
ax1.hist(data_normal, bins=30, density=True, alpha=0.7, color='skyblue')
ax1.plot(x_norm, y_norm, 'r-', lw=2)
ax1.set_title('Distribución Normal')
ax1.set_xlabel('Valores de X')
ax1.set_ylabel('Densidad')

# Gráfico para la distribución uniforme
ax2.hist(data_uniforme, bins=30, density=True, alpha=0.7, color='lightgreen')
ax2.plot(x_unif, y_unif, 'r-', lw=2)
ax2.set_title('Distribución Uniforme')
ax2.set_xlabel('Valores de X')
ax2.set_ylabel('Densidad')

plt.tight_layout()
plt.show()
```
